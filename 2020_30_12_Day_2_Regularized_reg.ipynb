```
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```
#creating the data 
m = 100
np.random.seed(42)
X = np.random.randn(m,1)
y = 0.5*X + np.random.randn(100,1)

sns.set()
plt.figure(figsize=(7,4))
plt.scatter(X,y)
plt.title("Data")
```

```
# we will first use the Linear regression 
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y.ravel())
X_new = np.array([[-2],[2]])
y_new = lin_reg.predict(X_new)
#plotting the best fit line 
plt.figure(figsize=(8,5))
plt.scatter(X,y)
plt.plot(X_new,y_new,"r-")
#finding the intercept and the coef 
print("coef :",lin_reg.coef_)
print("intercept :",lin_reg.intercept_)
```

```
#we will now add two outliers and see the effect on the linear regression 
max = X.argmax()
min = X.argmin()
y[max] =  25
y[min] = -25

#new data 
plt.figure(figsize=(8,5))
plt.scatter(X,y)
```

```
#fitting a linear regression in this data 
lin_reg1 = LinearRegression(normalize=True)
lin_reg1.fit(X,y)
X_new = np.array([[-2.6],[2]])
y_new = lin_reg1.predict(X_new)
#plotting the best fit line 
plt.figure(figsize=(8,5))
plt.scatter(X,y)
plt.plot(X_new,y_new,"r-")
```

```
#using the ridge regression to do the same 
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.5,solver='cholesky',normalize=True)
ridge_reg.fit(X,y)
X_new = np.array([[-2.6],[2]])
y_new = ridge_reg.predict(X_new)
plt.figure(figsize=(8,5))
plt.scatter(X,y)
plt.plot(X_new,y_new,"r-")
```

```
#using LAasso
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.5)
lasso_reg.fit(X,y)
X_new = np.array([[-2.6],[2]])
y_new = lasso_reg.predict(X_new)
plt.figure(figsize=(8,5))
plt.scatter(X,y)
plt.plot(X_new,y_new,"r-")
```

```
from sklearn.linear_model import ElasticNet
el_reg = ElasticNet(alpha=0.5,l1_ratio=0.5)
el_reg.fit(X,y)
X_new = np.array([[-2.6],[2]])
y_new = el_reg.predict(X_new)
plt.figure(figsize=(8,5))
plt.scatter(X,y)
plt.plot(X_new,y_new,"r-")
```
