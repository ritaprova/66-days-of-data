* Gradient Descent : This approach focuses towards finding the mvalue of the parameter function theta which helps in minimizing the cost function.

Important features of this approach include the stepsize which in thurn depends on the learning rate hyperparameter.

Note: the MSE cost function of a liner regression is a convex, and thus no local minima and only one global minima.

```
#creating the data 
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)
plt.figure(figsize=(8,5))
X = 2* np.random.rand(100,1)
y = 4 + 3*X + np.random.randn(100,1) #y=mx+c + noise
plt.scatter(X,y)
```

```
from sklearn.linear_model import SGDRegressor
reg2 = SGDRegressor(max_iter=1000,tol=1e-3,penalty=None,eta0=0.01)
reg2.fit(X,y.ravel())
print("the coef: ",reg2.coef_)
print("the intercept: ",reg2.intercept_)

#fitting the best fit line
#taking the 2 extreme values of the data 
x_new = np.array([[0],[2]])
y_new = reg2.predict(x_new)
#plotting 
sns.set()
plt.figure(figsize=(8,5))
plt.scatter(X,y)
plt.plot(x_new,y_new,"r-")
```
