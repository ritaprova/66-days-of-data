```
np.random.seed(42)
#creating some non linear data+noise
import numpy as np 
import pandas as pd 
m = 100
X = 6*np.random.rand(m,1) - 3
y = 0.5*X**2 + X + 2 + np.random.randn(m,1) #0.5x^2+x+2 + noise
#plotting 
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
import seaborn as sns
sns.set()
plt.scatter(X,y)

#fitting it with the help of the polynomial features transformer 
from sklearn.preprocessing import PolynomialFeatures
#creating a polynomial regressor pipeline 
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
poly_reg = make_pipeline(PolynomialFeatures(2),
                         LinearRegression())

#fitting the data
poly_reg.fit(X,y)
y_preds = poly_reg.predict(X)

plt.scatter(X,y_preds)
```

```
#using a transformer 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
poly_fit = PolynomialFeatures(degree=2,include_bias=True)
x = X.copy()
#transforming x
X_poly = poly_fit.fit_transform(x)
lin_reg = LinearRegression()
lin_reg.fit(X_poly,y)

#plotting the preds
y_pred = lin_reg.predict(X_poly)
plt.figure(figsize=(8,5))
plt.scatter(X,y)
plt.scatter(X,y_pred)
```
